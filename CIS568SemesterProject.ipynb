```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#make all of the different columns into numpy arrays for plotting
ID = df['ID'].to_numpy()

#bs is base soil
BSQ1Arr = df['BareSoilQ1'].to_numpy()
BSQ2Arr = df['BareSoilQ2'].to_numpy()
BSQ3Arr = df['BareSoilQ3'].to_numpy()

#gl is grasslands
GLQ1Arr = df['GrasslandsQ1'].to_numpy()
GLQ2Arr = df['GrasslandsQ2'].to_numpy()
GLQ3Arr = df['GrasslandsQ3'].to_numpy()

#p is poplar
PQ1Arr = df['PoplarQ1'].to_numpy()
PQ2Arr = df['PoplarQ2'].to_numpy()
PQ3Arr = df['PoplarQ3'].to_numpy()

#mf is mixed forest
MFQ1Arr = df['MixForestQ1'].to_numpy()
MFQ2Arr = df['MixForestQ2'].to_numpy()
MFQ3Arr = df['MixForestQ3'].to_numpy()
```

```python
#adding theme color for background
plt.figure(facecolor = "#0AB68B")

#plot the bacteria composition count against composition ID
plt.scatter(BSQ1Arr, ID, color = "#028174")

#add title and x axis label
plt.title("Bare Soil Quarter 1 Composition Scatterplot", fontweight = 'bold')
plt.xlabel("Count", fontweight = 'bold')

#use gca to define axes to color inside axes
ax = plt.gca()
ax.set_facecolor("#FFF0B6")

#label y axis
plt.ylabel("Bacteria Composition ID", fontweight = 'bold')

#save the figure
plt.savefig("scatterplotB1.png")
```

# Bar Charts

```python
#drop every row but the top three because the three bacteria compositions are being plotted
dfBar = df.drop(df.index[3:])
dfBar
```

```python
#make all of the different columns into numpy arrays for plotting
bcID = dfBar['ID'].to_numpy()

#bs is base soil
bcBSQ1Arr = dfBar['BareSoilQ1'].to_numpy()
bcBSQ2Arr = dfBar['BareSoilQ2'].to_numpy()
bcBSQ3Arr = dfBar['BareSoilQ3'].to_numpy()

#gl is grasslands
bcGLQ1Arr = dfBar['GrasslandsQ1'].to_numpy()
bcGLQ2Arr = dfBar['GrasslandsQ2'].to_numpy()
bcGLQ3Arr = dfBar['GrasslandsQ3'].to_numpy()

#p is poplar
bcPQ1Arr = dfBar['PoplarQ1'].to_numpy()
bcPQ2Arr = dfBar['PoplarQ2'].to_numpy()
bcPQ3Arr = dfBar['PoplarQ3'].to_numpy()

#mf is mixed forest
bcMFQ1Arr = dfBar['MixForestQ1'].to_numpy()
bcMFQ2Arr = dfBar['MixForestQ2'].to_numpy()
bcMFQ3Arr = dfBar['MixForestQ3'].to_numpy()
```

```python
#adding theme color for background
plt.figure(facecolor = "#0AB68B")

#make the bars
width = 0.2
plt.bar(bcID-0.3, bcBSQ1Arr, width, color='#2034ba')
plt.bar(bcID-0.1, bcGLQ1Arr, width, color='#b8911c')
plt.bar(bcID+0.1, bcPQ1Arr, width, color='#b3411b')
plt.bar(bcID+0.3, bcMFQ1Arr, width, color='#741bb3')

#add the title and x axis labels
plt.title("Quarter 1 Bacteria Compositions", fontweight = 'bold')
plt.xticks(bcID, ['Composition 1', 'Composition 2', 'Composition 3'])
plt.xlabel("Bacteria Compositions", fontweight = 'bold')

#use gca to define axes to color inside axes
ax = plt.gca()
ax.set_facecolor("#FFF0B6")

#set y label
plt.ylabel("Count", fontweight = 'bold')

#create the legend with the soil types
plt.legend(["Bare Soil", "Grasslands", "Poplar Plantation", "Mixed Forest"])

#save fig
plt.savefig("barchartQ1.png")
```
